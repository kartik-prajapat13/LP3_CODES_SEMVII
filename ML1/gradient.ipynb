Implement Gradient Descent Algorithm to find the local minima of a function. 
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.ta

```python
# Cell 2: Define the function and its gradient

# Function y = (x + 3)^2
def function(x):
    return (x + 3)**2

# Derivative of the function, dy/dx = 2 * (x + 3)
def gradient(x):
    return 2 * (x + 3)
```

```python
# Cell 3: Initialize parameters

# Start point
x_current = 2  # Starting at x = 2
learning_rate = 0.1  # Step size for each iteration
iterations = 100  # Number of iterations to converge
```

```python
# Cell 4: Implement Gradient Descent loop

# List to store x and y values for each iteration
x_values = []
y_values = []

for i in range(iterations):
    grad = gradient(x_current)  # Calculate the gradient at current x
    x_current = x_current - learning_rate * grad  # Update x by moving in the direction of the negative gradient
    y_current = function(x_current)  # Calculate the new y value for updated x
    
    # Append current x and y values to lists for tracking
    x_values.append(x_current)
    y_values.append(y_current)
    
    # Print the current state at each iteration
    print(f"Iteration {i+1}: x = {x_current}, y = {y_current}")

print(f"\nLocal minimum occurs at x = {x_current}, y = {y_current}")
```

```python
# Cell 5: (Optional) Visualize the gradient descent process using matplotlib

import matplotlib.pyplot as plt

# Plot y values over iterations to visualize convergence
plt.plot(range(iterations), y_values, marker='o', color='b')
plt.title("Gradient Descent Convergence")
plt.xlabel("Iteration")
plt.ylabel("Function Value y = (x + 3)^2")
plt.show()
```
